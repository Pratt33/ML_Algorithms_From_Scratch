```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y=make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.show()
```

```python
#let's apply ols first to find coefficient and intercept
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
plt.show()
```

```python
#time to apply gradient descent now assume slope is constant and m=78.35
#and let's b is starting from b=0
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='GD at b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
#let's ;earning rate 0.1
lr=0.1

step_size=loss_slope*lr
step_size
```

```python
#calculateing new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='green', label='b={}'.format(b))
plt.plot(X, y_pred, color='blue', label='b=0')
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred2, color='green', label='b={}'.format(b))
plt.plot(X, y_pred1, color='blue', label='b=0')
plt.plot(X, y_pred, color='yellow', label='b=0')
plt.legend()
plt.show()
```

```python
#iteration 3
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((78.35*X)+b).reshape(4)
plt.figure(figsize=(15, 15))
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred3, color='green', label='b={}'.format(b))
plt.plot(X, y_pred2, color='blue', label='b=0')
plt.plot(X, y_pred1, color='yellow', label='b=0')
plt.plot(X, y_pred, color='orange', label='b=0')
plt.legend()
plt.show()
```

```python
#now will do it with help of loop

#let's start bwith -100
b=-100
m=78.35
lr=0.1
for i in range(10):
    loss_slope=-2*np.sum(y-m*X.ravel()-b)
    step_size=loss_slope*lr
    b=b-step_size

    y_pred=m*X+b


    plt.plot(X, y_pred)

plt.scatter(X, y)
plt.show()
```

```python
#now will do it with help of loop
#with very samll lr
#let's start bwith -100
b=-100
m=78.35
lr=0.01
for i in range(10):
    loss_slope=-2*np.sum(y-m*X.ravel()-b)
    step_size=loss_slope*lr
    b=b-step_size

    y_pred=m*X+b


    plt.plot(X, y_pred)

plt.scatter(X, y)
plt.show()
```
